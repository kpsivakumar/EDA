### 1.1 Import required python libraries

```python
import pandas as pd

#for profiling report
import pandas_profiling
```

# Read Data File

```python
df = pd.read_csv('Sample_data.csv',index_col=[0])
```

# Pandas Profile Report

## Generate Pandas Profile Report

```python
pandas_profiling.ProfileReport(df)
```

## Save the pandas profile report as html file

```python
#save the pandas profiling report.
df.profile_report(title='Pandas Profiling Report').to_file(output_file="Student_Info_Pandas_profiling_Report.html")
```
